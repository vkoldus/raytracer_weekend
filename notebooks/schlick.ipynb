***Exploring Schlick's approximation***

$$
r_0=(\frac{1-r_i}{1+r_i})^2
$$

$$
R = r_0 + (1-r_0)*(1 - cos\theta)^5
$$

```python
import numpy as np
import matplotlib.pyplot as plt


def r0_func(refraction_index):
    return ((1 - refraction_index) / (1 + refraction_index)) ** 2


x = np.arange(0, 2, 0.1)
y1 = r0_func(x)
y2 = 1 - r0_func(x)

plt.figure()
plt.title("$r_0=(\\frac{1-r_i}{1+r_i})^2$")
plt.xlabel("$r_i$")
plt.ylabel("blue: $r_0$, green: $1-r_0$")
plt.plot(x, y1, "blue")
plt.plot(x, y2, "green")

x = np.arange(0, np.pi / 2, 0.1)
# y = np.cos(x)
# y = 1 - np.cos(x)
y = (1 - np.cos(x)) ** 5

plt.figure()
plt.title("$(1 - cos\\theta)^5$")
plt.xlabel("$\\theta$")
plt.plot(x, y)


def schlick(theta, r0):
    return r0 + (1 - r0) * np.power(1 - np.cos(theta), 5)


for ri in [1.33, 1.5]:
    r0 = r0_func(ri)
    print(r0)
    plt.figure()
    plt.title(f"Schlick for ri = {ri}")
    plt.xlabel("$\\theta$")
    plt.plot(x, schlick(x, r0))
```
